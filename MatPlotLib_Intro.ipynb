## Matplotlib

With Matplotlib you can create charts and graphs

First things first, we need to import the Matplotlib
into our Python environment

```python
# This will import and make them accessible with the shorter name plt
from matplotlib import pyplot as plt
```

### Basic Line Plot

Line graphs help when it comes to visualize how a variable changes 
over time.

Examples:

- Average prices of oil over a certain time
- Weight of a person over a certain time
- Average temperatures along a line of longitude over different latitudes

The following methods will create a simple line graph using `.plot()`
and display it with `.show()` :

```python
x_values = [0, 1, 2, 3, 4]
y_values = [0, 1, 4, 9, 16]
plt.plot(x_values, y_values)
plt.show()
```

`x_values` is a var holding a list of x-values for each point on the line graph

`y_values` is a var holding a list of y-values for each point on our line graph

`plt` is the abbreviation that we have given to the Matplotlib module

`plt.plot(x_values, y_values`) will create the line graph

`plt.show()` is the command to show the graph

### Example plotting lines from scratch

First we define two lists, in this example we want to help someone see their
spendings over the course of a week

```python
days = [0,1,2,3,4,5,6] # day 1 is 0 here, because a good plot should always start with 0
money_spent = [10,12,12,10,14,22,24]
# IMPORTANT: The number of x values MUST MATCH the number of y-values
```

```python
# Plot days on x-axis and money_spent on y-axis
plt.plot(days,money_spent)
# Visualize the plot
plt.show()
```

### Plot multiple lines

If we want to compare two datasets with the same scale and axis
categories we can have them displayed in the same plot.

Matplotlib will automatically place the two lines on the same axes
and give them different colors.

Lets have a look back at the data we have:

`days = [0, 1, 2, 3, 4, 5, 6]`

`money_spent = [10, 12, 12, 10, 14, 22, 24]` Money spent from person 1

`money_spent_2 = [11, 14, 15, 15, 22, 21, 12]` Money spent from person 2


```python
days = [0, 1, 2, 3, 4, 5, 6]
money_spent = [10, 12, 12, 10, 14, 22, 24]
money_spent_2 = [11, 14, 15, 15, 22, 21, 12]
# To show 2 graphs in one plot, call the plt.plot() twice !
plt.plot(days, money_spent)
plt.plot(days, money_spent_2)
plt.show()
# By default first line is always blue, second line is orange
# When you are plotting multiple lines, they do not need to have the
# same amount of x and y values, nor they need to share the same x values
```

### Axis and Labels

Sometimes when you want to highlight a specific area in your 
plot, it can help to zoom in or out to highlight some details.

To zoom, we can use `plt.axis()`.
We feed it a list as a input.

The list should contain:

- Minimum x-value displayed
- Maximum x-value displayed

- Minimum y-value displayed
- Maximum y-value displayed

Example:

Display a plot from `x=3` to `x=3` and
`y=2` to `y=5`

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.show()
```

```python
# The zoom-in
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.axis([0, 3, 2, 5])
plt.show()

# If you want, it is also possible for the axis to be decreasing in order as well
```

```python
# If you want, it is also possible for the axis to be decreasing in order as well
# This will draw the graph in reverse

y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.axis([3, 0, 5, 2])
plt.show()
```

### Label the Axes

`plt.ylabel()` Label y-axis

### Subplots

When you rather want to display two lines side-by-side
and not in the same set we would call each set of axes a subplot.

We can have many different subplots in the same figure and 
lay them out in different ways. Think of them as having rows & columns.

We can create subplots using `.subplot()`

The command `plt.subplot()` needs 3 arguments passed:

- Number of rows of subplots
- Number of columns of subplots
- Index of the subplot we want to. create

Any `plt.subplot()` that comes after a `plt.subplot()` will create
a line plot in the specified subplot.

```python
# Data sets
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
 
# First Subplot
plt.subplot(1, 3, 1)
plt.plot(x, y, color='green')
plt.title('First Subplot')
 
# Second Subplot
plt.subplot(1, 3, 2)
plt.plot(x, y, color='steelblue')
plt.title('Second Subplot')

# Third Subplot
plt.subplot(1,3,3)
plt.plot(x,y,'o', color='red') # the parameter 'o' creates a scatterplot!
plt.title('Third Subplot')
plt.subplots_adjust(wspace=1) # Create more space between plots
# Display both subplots
plt.show()
```

### Customize Subplots further

We have many options to further customize subplots, for example,
create some space between them, give them distinct titles & axis marking and more.

To do this, we use `.subplots_adjust()` in combination with some 
keyword arguments:

- `left` left-side margin, default value of 0.125 increase the number to make room for y-axis label

- `right` right-side margin, default value of 0.9, increase to make more room for figure, decrease to make room for legend

- `bottom` bottom margin, default value of 0.1, increase to make room for tick mark labels, or x-axis label

- `top` top margin, default value of 0.9

- `wspace` horizontal space between plots, default value of 0.2

- `hspace` vertical space between plots, default value of 0.2

Good explanation on [subplots](https://discuss.codecademy.com/t/how-can-rows-in-a-plot-have-different-numbers-of-columns/354492)

```python
# Apply this to the plots above, give them a bit more room horizontally

plt.subplots_adjust(wspace=0.35)
```

```python
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

plt.subplot(2,1,1)
plt.plot(x, straight_line)

plt.subplot(2,2,3)
plt.plot(x, parabola)

plt.subplot(2,2,4)
plt.plot(x, cubic)
plt.subplots_adjust(wspace=0.35,bottom=0.2)
plt.show()
```

### Legends

Multiple lines on a single graph can be labeled for better distinction
for that, we use the command `plt.legend()`

The legend method takes a list the labels to display

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'])
plt.show()
```

`plt.legend()` can also take a keyword argument which will define the position of the legend on the figure.

If you don't define `loc` , it will default to the "best position".


    0 best
    1 upper right
    2 upper left
    3 lower left
    4 lower right
    5 right
    6 center left
    7 center right
    8 lower center
    9 upper center
    10 center

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'], loc=6) # Will display legend "center left"
plt.show()
```

We can also use the `label` keyword inside our `plt.plot` then we don't need to give `plt.legend()` any 
arguments, but we still need to execute the command !

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16],
         label="parabola")
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64],
         label="cubic")
plt.legend() # Still need this command!
plt.show()
```

### Modify Ticks

To modify tick marks we have to specify which one we want to change.

To do that, we call `plt.subplot()` in a different way.

`ax = plt.subplot(1, 1, 1)`

`ax` is an axes object, it lets us modify the axes belonging to a specific subplot.

Even if we only have one subplot, when we want to modify its ticks,

we need to start by calling either `ax = plt.subplot(1, 1, 1)` or
`ax = plt.subplot()` in order to get our axes object.

For example we want ticks at x position 1,2 & 4, we would use something like 
the following:

```python
ax = plt.subplot()
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
ax.set_xticks([1, 2, 4])
plt.show()
```

Same thing for the y-ticks, we can modify them with `ax.set_yticks()`

When we change the ticks their labels automatically change to match.

But if we want special labels(such as strings), we can use

`ax.set_xticklabels()` & `ax.set_yticklabels()`

```python
ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8], 'o') # The 'o' argument creates a Scatterplot
ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', 'Test here', '80%'])
plt.show()
```

### Figures

When you plot a lot, it is easy to end up with lines that have been plotted but not displayed.

To clear memory, use `plt.close('all')` to clear all existing plots, before you create a new one.


The command `plt.figure()` creates a new figure and sizes them how we want.

With the added keyword `figsize=(width,height)` we can set the size of the figure(in inches).

To create a figure with a width of 4 inches and height of 10 inches:

```python
plt.figure(figsize=(4, 10))
plt.plot([1,3,7],[3,5,7])
plt.savefig('test_plot_picture.png')
plt.show()
```

This created a very skinny plot, we have want to save it for a presentation or to put 
it on a website, we can do this with `plt.savefig()`

With the added argument `png`,`svg` or `pdf` we can save it to a format we like.

`plt.savefig('name_of_graph.png')`

### Set limits for x and y axis to display

`plt.axis()` 

For example, a a display from `x=-5` to `x=5` and `y=0` to `y=10`
