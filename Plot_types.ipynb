# Different Plot Types

Different types of plots are good for different things,
here we go over some of the most common type of plots and learn
how and when to use them !

### Simple Bar Chart

`plt.bar` function allows you to create a simple bar chart to compare
multiple categories of data

Data that we could present with a bar chart:

- x-axis — famous buildings, y-axis — heights
- x-axis — different planets, y-axis — number of days in the year
- x-axis — programming languages, y-axis — lines of code written by you

We call `plt.bar`with two arguments:

- the x-values — a list of x-positions for each bar
- the y-values — a list of heights for each bar


In most cases our x-values should be a list like this `[0, 1, 2,...]`

It must have the same number of values as our `y-values list`

We can create this list manually, but also use a helper,
for example like this:

`heights = [88, 225, 365, 687, 4333, 10756, 30687, 60190, 90553]
x_values = range(len(heights))`

Here, the `range` function creates a list of consecutive integers.

The argument how many numbers should be in the list comes from `len(heights)`
which would be the exact number of elements we need.

```python
from matplotlib import pyplot as plt
```

```python
days_in_year = [88, 225, 365, 687, 4333, 10756, 30687, 60190, 90553]
plt.bar(range(len(days_in_year)),
        days_in_year)
plt.show()
```

Because there is no labeling, for anyone who just sees the plot,
it would be hard to grasp what is displayed here, we fix that later.

### Labeling the plot

Now we start to label a plot

Here analyze how good a drink from our fantasy coffee shop is selling

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales =  [91, 76, 56, 66, 52, 27]

plt.bar(range(len(drinks)), sales)
plt.show()
```

First as we learned before, we need to create an axes object
`ax = plt.subplot()`

Next we set the x-tick positions using a list of numbers
`ax.set_xticks([0,1,2,3,4,5])`

Then we label the ticks with a list of strings
`ax.set_xticklabels(["cappuccino", "latte", "chai", "americano", "mocha", "espresso"])`

With the extra `rotation` keyword we can rotate the label with a specified number of degrees.

`plt.xlabel` & `plt.ylabel` are also available to name the axes.

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales =  [91, 76, 56, 66, 52, 27]

ax = plt.subplot()
plt.bar(range(len(drinks)), sales)
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(["cappuccino", "latte", "chai", "americano", "mocha", "espresso"], rotation=30)
plt.xlabel('Coffees')
plt.ylabel('Amount of Sales')
plt.show()
```

### Side-By-Side Bars


We can use a bar char to compare two sets of data with
the same types of axis values.

Side-By-Side Bars can be good for:

- the populations of two countries over time

- prices for different foods at two different restaurants

- enrollments in different classes for males and females


### Example



```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

#Paste the x_values code here

n = 1  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store1_x = [t*element + w*n for element
             in range(d)]

plt.bar(store1_x, sales1)

n = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store2_x = [t*element + w*n for element
             in range(d)]

plt.bar(store2_x, sales2)

plt.show()
```

In the graph above, there are 6 sets of bars, with 2 bears in each set.
Each set has a width of 0.8(default value for all bars in Matplotlib).

- If our first blue bar is at x=0, then we want the next blue bar to be at x=2, and the next to be at x=4, etc

- Our first orange bar should be at x=0.8 (so that it is touching the blue bar), and the next orange bar should be at x=2.8, etc.


Python can do all this math for us with just a copy & paste of this code:

`n = 1  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store1_x = [t*element + w*n for element
             in range(d)]`
             
This generated the first set of x-values.

For the second set, we need this code again and make some changes:

`n = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store2_x = [t*element + w*n for element
             in range(d)]`
             

The code `[t*element + w*n for element in range(d)]` is called [list comprehension](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions)

Learn more [here](https://www.codecademy.com/articles/list-comprehension)

For side-by-side graphs you just need the code and define your `n`,
`t`, `d` & `w` correctly.

## Stacked Bars

Stacked Bars compare two data sets while preserve knowledge of the total
between them.



### Example

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

plt.bar(range(len(drinks)),sales1)
plt.bar(range(len(drinks)),sales2,bottom=sales1)
plt.legend(["Location 1","Location 2"])
plt.show()
```

As you can see, the first bar is plotted normally
`plt.bar(range(len(drinks)),sales1)`

and the second set of bars has `bottom` specified
`plt.bar(range(len(drinks)),sales2,bottom=sales1)`


## Error Bars

Sometimes we need to visually communicate some sort of uncertainty in
the heights of bars, examples are:

- The average number of students in a 3rd grade classroom is 30, but some classes have as few as 18 and others have as many as 35 students.

- We measured that the weight of a certain fruit was 35g, but we know that our scale isn’t very precise, so the true weight of the fruit might be as much as 40g or as little as 30g.

- The average price of a soda is 1.00 Dollar, but we also want to communicate that the standard deviation is 0.20 

To visually display them, we often use error bars to show where each bar 
could be taken errors into account.


### Example

With `yerr=2` we can show an error of +/- 2 to our `plot.bar()` command.
To make the caps wide an easier to read we can size
the caps with `capsize=10`

If we have a different error value for different bars, we can make
`yerr` equal to a list, rather than a single number.

Matplotlib reads this in the same order as the list of y-values.


```python
values = [10, 13, 11, 15, 20]
yerr = [1, 3, 0.5, 2, 4]
plt.bar(range(len(values)), values, yerr=yerr, capsize=10)
plt.show()
```

## Fill Between

We can not only display errors on bar charts, there is also a way 
to do it with graphs.

To do that, we use `plt.fill_between()` to visually shade error.

This function takes three arguments:

- `x-values` works just like x-values of `plt.plot()`
- lower-bound for y-values -sets the bottom of the shaded area
- upper-bound for y-values - sets the top of the shaded area


We use `.fill_between()` to create a shaded error region and then
plot the actual line over it.

The `alpha` keyword can have a value between 0 and 1.
It stands for the transparency so that we can see the line underneath.

### Example:

This example shows data with a error of +/- 2:

```python
x_values = range(10)
y_values = [10, 12, 13, 13, 15, 19, 20, 22, 23, 29]
y_lower = [8, 10, 11, 11, 13, 17, 18, 20, 21, 27]
y_upper = [12, 14, 15, 15, 17, 21, 22, 24, 25, 31]
 
plt.fill_between(x_values, y_lower, y_upper, alpha=0.2) #this is the shaded error
plt.plot(x_values, y_values) #this is the line itself
plt.show()
```

Having to calculate `y_lower` and & `y_upper` by hand is time-consuming.

In order to correctly add or substract from a list, we need
to use [list comprehension](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions)

`y_lower = [i - 2 for i in y_values]` substracts 2 from every y value

`y_upper = [i + 2 for i in y_values]` adds 2 to every v value

### Example 2:

```python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]

#your work here
y_lower = [i - 0.1 * i for i in revenue]
y_upper = [i + 0.1 * i for i in revenue]
plt.fill_between(months, y_lower, y_upper, alpha=0.2)
plt.plot(months, revenue)
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)

plt.show()
```

## Pie Charts


When we want to display elements of a data set as proportions of a whole,
we can use a pie chart.

Pie charts are useful for displaying data like:

- Different ethnicities that make up a school district
- Different macronutrients (carbohydrates, fat, protein) that make up a meal
- Different responses to an online poll

In Matplotlib you can create a pie chart with the command `plt.pie` and pass
in the values you want to chart.

If the chart looks weird and tilted we can fix this with
`plt.axis('equal')`

```python
budget_data = [500, 1000, 750, 300, 100]
 
plt.pie(budget_data)
plt.axis('equal')
plt.show()
```

### Label Pie Charts

There are two ways to understand what each slice of the pie represents

- use a legend to label each color
- put labels on the chart itself

Legend method:

```python
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']
 
plt.pie(budget_data)
plt.legend(budget_categories)
plt.show()
```

This will put the category names into a legend on the chart.


Put labels on the chart itself:

```python
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']

plt.pie(budget_data, labels=budget_categories)
plt.show()
```

### Add percentage of the total to the slices

Matplotlib can do this automatically with `autocpt`
We pass in [string formatting instructions](https://docs.python.org/2/library/stdtypes.html#string-formatting) to
format the labels how we want.

Examples:

`%0.2f` 2 decimal places, like 6.08

`%0.2f%%` 2 decimal places, but with a percent sign at the end, like 6.08%(you need two %
at the end, because the first acts as an escapce character)

`%d%%` rounded to the nearest int, with an percent sign at the end, like 6%

```python
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']

plt.pie(budget_data,
        labels=budget_categories,
        autopct='%0.1f%%')
plt.show()
```
